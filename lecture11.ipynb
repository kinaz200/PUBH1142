### Imports

```python
# Imports
import numpy as np
from scipy import stats
```

### create population 

```python
# Seed the pseudo-random number generator
np.random.seed(28)

# Create population (diastolic blood pressure)
dBP = np.random.normal(loc=85, scale=10, size=100000)

# Population mean
population_mean = dBP.mean()
population_mean
```

### sampling 

```python
# Set new seed
np.random.seed(12)

# Take random sample of 40
sample = np.random.choice(dBP, size=40, replace=False)

sample
```

### statistics

```python
# Sample mean
sample_mean = sample.mean()

# Sample standard deviation (ddof=1 for sample std)
sample_std = sample.std(ddof=1)

# Standard error
standard_error = sample_std / np.sqrt(len(sample))

sample_mean, sample_std, standard_error
```

### T-value 

```python
# Degrees of freedom
df = len(sample) - 1

# 95% confidence level → alpha = 0.05
c = stats.t.ppf(1 - 0.025, df)

c
```

### Confidence Interval

```python
# Lower bound
lower_bound = sample_mean - c * standard_error

# Upper bound
upper_bound = sample_mean + c * standard_error

lower_bound, upper_bound
```

### Bootstrap Sampling 

```python
# Bootstrap means
bootstrapped_means = []

for _ in range(1000):
    boot_sample = np.random.choice(sample, size=40, replace=True)
    bootstrapped_means.append(boot_sample.mean())

# Convert to array and sort
bootstrapped_means = np.array(bootstrapped_means)
bootstrapped_means.sort()

bootstrapped_means[:10]  # preview
```

### Percents 

```python
# 2.5th percentile
p2_5 = np.percentile(bootstrapped_means, 2.5)

# 97.5th percentile
p97_5 = np.percentile(bootstrapped_means, 97.5)

p2_5, p97_5
```
