# Health Data Ethics Assignment

## Misuse of Data

A well-known example of health data misuse is the case of **Henrietta Lacks**. In 1951, doctors took her cancer cells without her consent and used them for research, creating the HeLa cell line that led to major medical breakthroughs. Neither Lacks nor her family were told or compensated, raising serious ethical issues around informed consent, exploitation, and privacy, especially when her genetic data was later shared.  

This could have been prevented through clear consent, transparency, and respect for patient autonomy. Like the Havasupai case, it shows how misusing health data can harm trust, even when the research benefits society.

**Source:**  
https://pmc.ncbi.nlm.nih.gov/articles/PMC3516052/

## Breaches in Health Data Science

The **Premera Blue Cross data breach in 2015** exposed the health and personal data of over 11 million people. Information like Social Security numbers and medical records was put at risk, which left many people worried about identity theft and financial fraud.  

Premera faced lawsuits and a major investigation by state attorneys general and was ordered to pay about $10 million for failing to protect people’s data and for not being honest about what happened. Washington’s share of that settlement went toward stronger data privacy enforcement and forced Premera to improve how it protects data with better systems and training.  

Many people were also offered credit monitoring and identity theft protection because of the breach. Beyond the legal and financial consequences, the incident damaged trust in how health data is handled and made people more hesitant to share sensitive information needed for care or research.

https://www.atg.wa.gov/news/news-releases/attorney-general-ferguson-s-investigation-premera-data-breach-results-premera

## Bias in Health Data

An example of bias in health data is the use of a health care algorithm in the US that decides who should get extra care based on past health care costs. Because Black patients have historically had less access to care, they often had lower costs even when they were just as sick as white patients. As a result, the system ranked them as lower risk and they missed out on needed care.  

The consequence was that existing health gaps became worse and trust in the health system was damaged. This could have been prevented by using actual medical need instead of cost data and by testing the system for bias before it was used.

**Source:**  
https://learn.hms.harvard.edu/insights/all-insights/confronting-mirror-reflecting-our-biases-through-ai-health-care

## Anonymization of Health Data

One example of anonymization in health data comes from research on sharing electronic health records such as clinical notes and discharge summaries. In projects analyzed by NIH researchers, patient identifiers are removed or replaced with semantically similar surrogate values so that the data can be shared without revealing who the patient actually is.  

For instance, real names, dates of birth, and other direct identifiers are taken out or swapped. More advanced methods like document clustering and value enumeration are used to group similar records and then mask quasi-identifiers such as age, gender, or ZIP code at the group level so individuals cannot be re-identified but the medical information remains useful for research.  

This method goes beyond simply deleting names and numbers by also handling text in medical notes that might indirectly reveal identity, making the data safer to share while keeping it meaningful for analysis.

**Source:**  
https://pmc.ncbi.nlm.nih.gov/articles/PMC5858761/

# Quiz

**1. What is the primary goal of data anonymization?**  
To protect people’s privacy while still allowing data to be used for research.

**2. What is the Havasupai Tribe case about?**  
Researchers used tribe members’ blood samples for studies they never agreed to, breaking trust and informed consent.

**3. What was the largest health data breach in history?**  
The Anthem breach in 2015, which exposed the data of about 78 million people.

**4. What is the pulse oximeter controversy about?**  
The devices are less accurate on people with darker skin, which can delay care and cause harm.

**5. What is the role of an Institutional Review Board (IRB) in research?**  
To protect research participants by reviewing studies for ethics, consent, and safety.

**6. What is the difference between spreadsheet software and database software for health data management?**  
Spreadsheets are simple and manual, while databases handle large data, multiple users, and better security.

**7. What is a conflict of interest in public health research?**  
When a researcher’s personal or financial interests could affect their decisions or results.

**8. What is informed consent in the context of health data collection?**  
When people are fully told how their data will be used and agree to it.

**9. What is data masking in the context of de-identification of health data?**  
Hiding or replacing personal details so individuals cannot be identified.

**10. What is the principle of k-anonymity?**  
Each person’s data looks the same as at least k other people, making them harder to identify.

**11. What is the main concern about conflicts of interest in public health research?**  
They can bias results and hurt public trust.

**12. What is the first step in managing conflicts of interest?**  
Disclosing them.

**13. What is the role of independent oversight in managing conflicts of interest?**  
To review research objectively and keep it fair.

**14. What is one strategy for mitigating conflicts of interest?**  
Limiting the role of people with conflicts in decision making.

**15. What is the potential impact of not properly managing conflicts of interest?**  
Biased research, bad policies, and loss of public trust.
